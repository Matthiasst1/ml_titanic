```python
import pandas as pd
import numpy as np

train_data=pd.read_csv('train.csv')
train_data.head()
```

```python
test_data=pd.read_csv('test.csv')
test_data.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

fig, ax = plt.subplots()

### bam ist die Variable für SibSp = set {0 bis 8} 
for bam in set(train_data['SibSp']):
    age_fare = train_data[train_data.SibSp == bam][['Age', 'Fare']].dropna()
    print(str(bam) + " Siblings: "+ str(len(age_fare)))
    age = age_fare["Age"]
    fare = age_fare['Fare']
    colors = cm.rainbow(np.linspace(0, 1, len(age_fare)))
    ax.scatter(list(age), list(fare), marker= ".", s=20, color=colors)

plt.ylabel('Price for Ticket in $', fontsize=11)
plt.xlabel('Age of passenger', fontsize=11)
plt.style.use('dark_background')
plt.show() 
```
